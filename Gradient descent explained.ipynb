```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gradient_descent(x,y):
    # m_curr = initial gradient, b_current = initial bias
    m_curr = b_curr = 0
    
    # l_rate = learning rate
    l_rate = 0.01
    
    n = len(x)
    plt.scatter(x,y,color='red',marker='+',linewidth='8')
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
        # print (m_curr,b_curr, i)
        plt.plot(x,y_predicted,color='green')
        # md = derivative of the squared error wrt m the initial slope
        md = -(2/n)*sum(x*(y-y_predicted))
        
        # yd = derivative of the squared error wrt the predicted y
        yd = -(2/n)*sum(y-y_predicted)
        
        m_curr = m_curr - l_rate * md
        b_curr = b_curr - l_rate * yd
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
